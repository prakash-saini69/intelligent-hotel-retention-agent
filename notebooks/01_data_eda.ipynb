```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Connect to the Database
# NOTE: We use ".." to go back up one level from 'notebooks' to the root, then into 'data'
db_path = "../data/hotel_retention.db"
conn = sqlite3.connect(db_path)
```

```python
# 2. Load Data
query = "SELECT * FROM bookings"
df = pd.read_sql(query, conn)

print("‚úÖ Data Loaded Successfully!")
print(f"Total Rows: {len(df)}")
display(df.head())
```

```python
# 3. Quick Analysis of "High Risk" Candidates
# Let's filter for people who might be risky (e.g., High Price + History of cancellations)

print("\n--- üîç Potential High Risk Customers ---")
high_risk_filter = (df['previous_cancellations'] > 0) & (df['booking_price'] > 500)
risky_customers = df[high_risk_filter]

if not risky_customers.empty:
    display(risky_customers[['customer_id', 'name', 'booking_price', 'previous_cancellations']])
else:
    print("No high-risk customers found with current simple filter.")
```

```python
# 4. Define 'Churn' Logic (For ML later)
# We don't have a 'cancelled' column for future bookings yet, 
# but for training data, we usually look at 'status'.
# Let's check value counts of status.
print("\n--- üìä Booking Status Distribution ---")
print(df['status'].value_counts())
```
